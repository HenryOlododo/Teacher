```python
import matplotlib.pyplot as plt

# Function to create a bar chart
def create_bar_chart(data, title, xlabel, ylabel, filename):
    categories = list(data.keys())
    frequencies = list(data.values())
    
    plt.figure(figsize=(10, 6))
    plt.bar(categories, frequencies, color='#4CAF50', edgecolor='black')
    plt.title(title, fontsize=14, pad=20)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.savefig(filename)
    plt.show()

# Section A: Demographic Variables
# Age Range
age_data = {'26-30': 20, '30-35': 20, '36-40': 10, '36-47': 28, '48 and above': 28}
create_bar_chart(age_data, 'Section A: Age Range Distribution', 'Age Range', 'Frequency', 'age_range.png')

# Gender
gender_data = {'Male': 68, 'Female': 38}
create_bar_chart(gender_data, 'Section A: Gender Distribution', 'Gender', 'Frequency', 'gender.png')

# Marital Status
marital_data = {'Single': 20, 'Married': 86}
create_bar_chart(marital_data, 'Section A: Marital Status Distribution', 'Marital Status', 'Frequency', 'marital_status.png')

# Religion
religion_data = {'Christianity': 68, 'Muslim': 38}
create_bar_chart(religion_data, 'Section A: Religion Distribution', 'Religion', 'Frequency', 'religion.png')

# Ethnicity
ethnicity_data = {'Yoruba': 68, 'Igbo': 28, 'Hausa': 10}
create_bar_chart(ethnicity_data, 'Section A: Ethnicity Distribution', 'Ethnicity', 'Frequency', 'ethnicity.png')

# Education
education_data = {'Tertiary': 106}
create_bar_chart(education_data, 'Section A: Education Distribution', 'Education Level', 'Frequency', 'education.png')

# Section B: Agreement Statements on Physical Assessment
# Combine all statements into one chart for simplicity
statement_data = {
    'Stmt 1 (Comprehensive)': [106, 0],
    'Stmt 2 (Entire Body)': [106, 0],
    'Stmt 3 (Early Detection)': [106, 0],
    'Stmt 4 (Vision/Hearing)': [68, 38],
    'Stmt 5 (Privacy Rooms)': [48, 58]
}
fig, ax = plt.subplots(figsize=(12, 6))
x = range(len(statement_data))
agree = [data[0] for data in statement_data.values()]
disagree = [data[1] for data in statement_data.values()]
width = 0.35
ax.bar([i - width/2 for i in x], agree, width, label='Agree', color='#4CAF50', edgecolor='black')
ax.bar([i + width/2 for i in x], disagree, width, label='Disagree', color='#FF5733', edgecolor='black')
ax.set_title('Section B: Agreement on Physical Assessment Statements', fontsize=14, pad=20)
ax.set_xlabel('Statements', fontsize=12)
ax.set_ylabel('Frequency', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(statement_data.keys(), rotation=45, ha='right')
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('section_b_statements.png')
plt.show()

# Section C: Techniques and Practices
# Combine practices into one chart
practice_data = {
    'P1 (Vision/Smell/Hearing)': [10, 96, 0],
    'P2 (Touch/Feel)': [0, 96, 10],
    'P3 (Tap Body)': [0, 20, 86],
    'P4 (Privacy)': [68, 28, 10],
    'P5 (Prof. Dev.)': [58, 28, 20],
    'P6 (Stay Updated)': [96, 10, 0],
    'P7 (Report Outcomes)': [86, 20, 0],
    'P8 (Collaborate)': [96, 10, 0]
}
fig, ax = plt.subplots(figsize=(14, 6))
x = range(len(practice_data))
always = [data[0] for data in practice_data.values()]
sometimes = [data[1] for data in practice_data.values()]
never = [data[2] for data in practice_data.values()]
width = 0.25
ax.bar([i - width for i in x], always, width, label='Always', color='#4CAF50', edgecolor='black')
ax.bar(x, sometimes, width, label='Sometimes', color='#FFC107', edgecolor='black')
ax.bar([i + width for i in x], never, width, label='Never', color='#FF5733', edgecolor='black')
ax.set_title('Section C: Techniques and Practices', fontsize=14, pad=20)
ax.set_xlabel('Practices', fontsize=12)
ax.set_ylabel('Frequency', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(practice_data.keys(), rotation=45, ha='right')
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('section_c_practices.png')
plt.show()

# Section D: Challenges in Physical Assessment
# Combine challenges into one chart
challenge_data = {
    'C1 (Training)': [68, 38],
    'C2 (Prof. Support)': [78, 28],
    'C3 (Time)': [48, 58],
    'C4 (Tools)': [58, 48],
    'C5 (Private Space)': [0, 106],
    'C6 (Consent)': [68, 38],
    'C7 (Uncertainty)': [68, 38],
    'C8 (Benchmark)': [106, 0],
    'C9 (Uncooperative)': [68, 38],
    'C10 (Documentation)': [68, 38]
}
fig, ax = plt.subplots(figsize=(14, 6))
x = range(len(challenge_data))
agree = [data[0] for data in challenge_data.values()]
disagree = [data[1] for data in challenge_data.values()]
width = 0.35
ax.bar([i - width/2 for i in x], agree, width, label='Agree', color='#4CAF50', edgecolor='black')
ax.bar([i + width/2 for i in x], disagree, width, label='Disagree', color='#FF5733', edgecolor='black')
ax.set_title('Section D: Challenges in Physical Assessment', fontsize=14, pad=20)
ax.set_xlabel('Challenges', fontsize=12)
ax.set_ylabel('Frequency', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(challenge_data.keys(), rotation=45, ha='right')
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('section_d_challenges.png')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for Section A: Demographic Variables
demographic_data = {
    'Age Range': {'26-30': 20, '30-35': 20, '36-40': 10, '36-47': 28, '48 and above': 28},
    'Gender': {'Male': 68, 'Female': 38},
    'Marital Status': {'Single': 20, 'Married': 86},
    'Religion': {'Christianity': 68, 'Muslim': 38},
    'Ethnicity': {'Yoruba': 68, 'Igbo': 28, 'Hausa': 10},
    'Education': {'Tertiary': 106}
}

# Create a figure with subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10), constrained_layout=True)
axes = axes.flatten()  # Flatten the 2x3 grid for easier iteration

# Plot each demographic variable
for idx, (title, data) in enumerate(demographic_data.items()):
    categories = list(data.keys())
    frequencies = list(data.values())
    axes[idx].bar(categories, frequencies, color='#4CAF50', edgecolor='black')
    axes[idx].set_title(title, fontsize=12, pad=10)
    axes[idx].set_xlabel(title, fontsize=10)
    axes[idx].set_ylabel('Frequency', fontsize=10)
    axes[idx].tick_params(axis='x', rotation=45, labelsize=8)
    axes[idx].grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout and save
plt.suptitle('Section A: Demographic Variables Distribution (N=106)', fontsize=16, y=1.05)
plt.savefig('section_a_demographics.png')
plt.show()
```
